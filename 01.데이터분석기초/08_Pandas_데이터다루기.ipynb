**예제 01**

다음과 같이 연도, 월, 날짜가 기입된 데이터프레임을 작성하고 `df01` 로 바인딩하세요.

2001년 1-1일 부터 2001년 2-28일까지 기입되어 있습니다. (31+28= 총 59줄)


| index | 연도 | 날짜 | 
|--|--|--|
|0|2001|1-1|
|1|2001|1-2|
|2|2001|1-3|
||...|

````python
df01 = pd.DataFrame({ '연도' : [2001], '날짜' : ['1-1']})
for i in range(1) : 
  for j in range(31) :
    df01.loc[i + j] = [2001 , f'{i+1}-{j+1}']

for i in range (28) : 
  df01.loc[31 + i] = [2001, f'2-{i+1}']

df01.head(), df01.tail()
````

**예제 02**

`df01`의 연도를 모두 2022로 바꾸세요.

````python
df01['연도'] = 2022
df01.head()
````

**예제 03**

`df01` 에 1월, 2월에 이어서 3월을 추가하세요.

````python
for i in range (31) : 
  df01.loc[59 + i] = [2022, f'3-{i+1}']

df01.tail()
````

**예제 05**

**예제 06**
