      * origin: 제조 국가
      * name: 이름
  * 모델링을 위한 자료형 구분
    * 연속 -> ...
    * 이산 -> ...

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# 시간이 오래 걸리므로 소량의 샘플을 그리는게 좋다.

# df.sample(100) # 데이터 프레임에서 샘플 100개 추출
sns.pairplot(df.sample(100), hue = 'origin')
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
# x = 각 범주 자료형 y = mpg 연비
# nums 리스트
f, axes = plt.subplots(1, len(noms), figsize=(12,4))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.boxplot(data=df, x=col, y = 'mpg', ax=ax)
plt.show()
```

* 최신 모델일 수록 연비가 좋아진다

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
# 연속형과 연속형 비교
# x = 연속 자료형, y = mpg
# nums 리스트의 자료형을 그린다.
f, axes = plt.subplots(1, len(nums), figsize=(15,6))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.scatterplot(data=df, x=col, y = 'mpg', ax=ax)
plt.show()
```

* `hue` 옵션에 `origin` 추가해서 그리기 

```python
# 연속형과 연속형 비교
# x = 연속 자료형, y = mpg
# nums 리스트의 자료형을 그린다.
f, axes = plt.subplots(1, len(nums), figsize=(15,6))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.scatterplot(data=df, x=col, y = 'mpg', ax=ax, hue = 'origin')
plt.show()
```

* `regplot`으로 회귀 시각화

```python
f, axes = plt.subplots(1, len(nums), figsize=(15,6))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.regplot(data=df, x=col, y = 'mpg', ax=ax, marker= '.')
plt.show()
```

* `residplot`으로 잔차 시각화

```python
f, axes = plt.subplots(1, len(nums), figsize=(15,6))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.residplot(data=df, x=col, y = 'mpg', ax=ax)
plt.show()
```

* `lmplot`으로 `origin`별 회귀 시각화

```python
# f, axes = plt.subplots(1, len(nums), figsize=(15,6))
# axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.lmplot(data=df, x=col, y = 'mpg', hue = 'origin')
  plt.show()
```
