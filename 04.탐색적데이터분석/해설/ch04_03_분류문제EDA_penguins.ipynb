## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
f, axes = plt.subplots(1, len(nums), figsize=(20,3))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.histplot(data=df, x=col, ax=ax)
plt.show()
```

* 이산 자료형 분포 그리기

```python
f, axes = plt.subplots(1, len(noms), figsize=(15,3))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.countplot(data=df, x=col, ax=ax)
plt.show()
```

### | **기술 통계량 시각화**

* 기술통계량 확인

```python
df.describe()
```

* boxplot 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1, len(nums), figsize=(20,3))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.boxplot(data=df, x=col, ax=ax)
plt.show()
```

* viloinplot 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1, len(nums), figsize=(20,3))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.violinplot(data=df, x=col, ax=ax)
plt.show()
```

### **| 예측값에 대한 추가 시각화**

* `countplot`으로 그리기


```python
_=sns.countplot(data=df, x=y)
```

* `hue` 옵션에 범주형 속성 추가하기

```python
f, axes = plt.subplots(1,2, figsize=(8,3))
axes = axes.flatten()
for nom, ax in zip(noms, axes):
  sns.countplot(data=df, x=y, ax=ax, hue=nom)
plt.show()
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
_=sns.pairplot(df.sample(100), hue=y)
```

* `kdeplot` 이용해 연속 자료형 별 예측값의 분포 시각화 하기

```python
for col in nums:
  sns.displot(data=df, x=col, hue=y, kind='kde')
  plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `island` 추가하기

```python
for col in nums:
  sns.displot(data=df, x=col, hue=y, kind='kde', col='island')
  plt.show()
```

* `bill_length_mm`과 `bill_depth_mm`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
_=sns.scatterplot(data=df,x=nums[0], y=nums[1], hue=y)
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `island` 추가하기

```python
_=sns.lmplot(data=df,x=nums[0], y=nums[1], hue=y, col='island')
```

* `flipper_length_mm`, `body_mass_g`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
_=sns.scatterplot(data=df,x=nums[2], y=nums[3], hue=y)
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `island` 추가하기

```python
_=sns.lmplot(data=df,x=nums[2], y=nums[3], hue=y, col='island')
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
f, axes = plt.subplots(1,2, figsize=(10,4))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.histplot(data=df, x=col, y=y,ax = ax)
plt.show()
```
