```
from sklearn.datasets import load_iris
iris = load_iris()
```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()

x_names = iris.feature_names
print(x_names)
y_names = iris.target_names
print(y_names)

xs = iris.data
print('x',xs.shape)
y = iris.target
print('y',y.shape)
````

````python
plt.scatter(xs[:,0], xs[:,1], c=y, cmap='rainbow')
plt.title('Species by sepal size')
plt.xlabel(x_names[0])
plt.ylabel(x_names[1])
plt.show()
````

````python
plt.scatter(xs[:,2], xs[:,3], c=y, cmap='rainbow')
plt.title('Species by petal size')
plt.xlabel(x_names[2])
plt.ylabel(x_names[3])
plt.show()
````

````python
plt.scatter(xs[:,2], xs[:,3], s=xs[:,1]**4, c=y, 
            cmap=plt.cm.get_cmap('rainbow',3), alpha=0.5)
plt.title('Species by petal size')
plt.xlabel(x_names[2])
plt.ylabel(x_names[3])
plt.colorbar(ticks=[0,1,2], label='Species')
plt.clim(-.5,2.5)
plt.show()
````

  * `x`축은 `latitude` 입니다.
  * `y`축은 `longitude` 입니다.
  * 점의 크기는 `population`에 비례합니다.
  * `median_house_value`로 색깔을 나타내고 `colorbar`를 추가합니다.
* 색과 마커, 비례 정도 등 세세한 부분은 각자의 스타일로 그리세요.

![](https://github.com/yebiny/SkillTreePython-DataAnalysis/blob/main/imgs/ch0205-01.png?raw=true)

````python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df.head()
````

````python
plt.figure(figsize=(12,8))
plt.scatter(df.latitude, df.longitude, 
            s=df.population/20, 
            c=df.median_house_value/1000,
            alpha=0.5,
            cmap='rainbow')
plt.xlabel('Latitude', fontsize=12)
plt.ylabel('Longitude', fontsize=12)
plt.title('California house value', fontsize=18)
cb = plt.colorbar(label='Median house value', shrink=.8 )
plt.show()
````
